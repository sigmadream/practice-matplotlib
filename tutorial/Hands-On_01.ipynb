```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Plotting one value on a single graph

```python
dev_x = [x for x in range(25,36,1)]
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(dev_x, dev_y)
plt.title("Median Salary (USD) by Age")
plt.xlabel("Age")
plt.ylabel("Median Salary (USD)")
```

## Plotting two values on a single graph

```python
age_x = dev_x
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
pydev_y = [45372, 48476, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
jsdev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
```

```python
plt.plot(age_x, dev_y)
plt.plot(age_x, pydev_y)
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
```

```python
# Tip1: 플롯에 추가된 순서대로 목록을 전달하여 범례를 추가할 수 있습니다.
plt.plot(age_x, dev_y)
plt.plot(age_x, pydev_y)
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend(["All_Developers", "Python_Developers"])
```

```python
# Tip2: plt.plot()에 레이블 매개변수 추가하기
plt.plot(age_x, dev_y, label = "All_Developers")
plt.plot(age_x, pydev_y, label = "Python_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
```

```python
plt.plot(age_x, dev_y, "k--", label = "All_Developers")
plt.plot(age_x, pydev_y, "b-",label = "Python_Developers")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
# 코드 가독성 높이기
plt.plot(age_x, dev_y, color = "k", linestyle = "--", label = "All_Developers")
plt.plot(age_x, pydev_y, color = "b", linestyle = '-', label = "Python_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
```

```python
# 선에 마커 추가하기
plt.plot(age_x, dev_y, color = "k", linestyle = "--", marker = ".", label = "All_Developers")
plt.plot(age_x, pydev_y, color = "b", linestyle = '-', marker = "o",label = "Python_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
```

```python
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',label = "Python_Developers")
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',label = "Java_script_Developers")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
# 파이썬과 자바스크립트의 너비 변경하기
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',linewidth = 4, label = "Python_Developers")
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',linewidth = 4, label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
# 그래프에 패딩 추가
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',linewidth = 4, label = "Python_Developers")
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',linewidth = 4, label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
plt.tight_layout()
```

```python
# 플롯에 그리드 추가
plt.plot(age_x, pydev_y, color = "#5a7d9a", linestyle = '-',linewidth = 4, label = "Python_Developers")
plt.plot(age_x, jsdev_y, color = "#adad3b", linestyle = '-',linewidth = 4, label = "Java_script_Developers")
plt.plot(age_x, dev_y, color = "#444444", linestyle = "--", label = "All_Developers")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
plt.tight_layout()
plt.grid(True)
```

## Line and Fill!

```python
data = pd.read_csv("data/salary.csv")
data.head()
```

```python
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]
```

```python
plt.style.use("ggplot")
plt.plot(ages, dev_salaries, color = "#444444", linestyle = '--', label = "All Devs")
plt.plot(ages, py_salaries, label = "Python")
plt.legend()
plt.tight_layout()
```

```python
# 채우기
plt.style.use("ggplot")
plt.plot(ages, dev_salaries, color = "#444444", linestyle = '--', label = "All Devs")
plt.plot(ages, py_salaries, label = "Python")
plt.fill_between(ages, py_salaries)
plt.legend()
plt.tight_layout()
```

```python
plt.style.use("ggplot")
plt.plot(ages, dev_salaries, color = "#444444", linestyle = '--', label = "All Devs")
plt.plot(ages, py_salaries, label = "Python")
plt.fill_between(ages, py_salaries, alpha = 0.5)
plt.legend()
plt.tight_layout()
```

```python
overall_median = 57287
plt.style.use("ggplot")
plt.plot(ages, dev_salaries, color = "#444444", linestyle = '--', label = "All Devs")
plt.plot(ages, py_salaries, label = "Python")
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate = True, alpha = 0.5)
plt.legend()
plt.tight_layout()
```

```python
overall_median = 57287
plt.style.use("ggplot")
plt.plot(ages, dev_salaries, color = "#444444", linestyle = '--', label = "All Devs")
plt.plot(ages, py_salaries, label = "Python")
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate = True, alpha = 0.5)
plt.fill_between(ages, py_salaries, overall_median, where = (py_salaries < overall_median), interpolate = True, alpha = 0.25)
plt.legend()
plt.tight_layout()
```

```python
# 채우기 색상 사용자 지정
plt.style.use("ggplot")
plt.plot(ages, dev_salaries, color = "#444444", linestyle = '--', label = "All Devs")
plt.plot(ages, py_salaries, label = "Python")
overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate = True, alpha = 0.5)
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries < overall_median), color = "green", interpolate = True, alpha = 0.25)
plt.legend()
plt.tight_layout()
```

```python
overall_median = 57287
plt.style.use("ggplot")
plt.plot(ages, dev_salaries, color = "#444444", linestyle = '--', label = "All Devs")
plt.plot(ages, py_salaries, label = "Python")
plt.fill_between(ages, py_salaries, dev_salaries, where = (py_salaries > dev_salaries), interpolate = True, alpha = 0.5)
plt.fill_between(ages, py_salaries, overall_median, where = (py_salaries < dev_salaries), color = "green", interpolate = True, alpha = 0.25)
plt.legend()
plt.tight_layout()
```

```python
overall_median = 57287
plt.style.use("ggplot")
plt.plot(ages, dev_salaries, color = "#444444", linestyle = '--', label = "All Devs")
plt.plot(ages, py_salaries, label = "Python")
plt.fill_between(ages, py_salaries, dev_salaries, where = (py_salaries > dev_salaries), interpolate = True, alpha = 0.5, label = "above dev_sal")
plt.fill_between(ages, py_salaries, overall_median, where = (py_salaries < dev_salaries), color = "green", interpolate = True, alpha = 0.25, label = "below dev_sal")
plt.legend()
plt.tight_layout()
```
